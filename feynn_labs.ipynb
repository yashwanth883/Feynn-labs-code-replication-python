```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate synthetic data
df, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Apply k-means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(df)

# Get cluster centroids and labels
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

# Plot the data points and cluster centroids
plt.scatter(df[:, 0], df[:, 1], c=labels, cmap='viridis', edgecolors='k', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.utils import resample
from sklearn.metrics import adjusted_rand_score
from sklearn.datasets import make_blobs

# Set seed for reproducibility
np.random.seed(1234)

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Step 1: K-means clustering with different numbers of clusters
cluster_results = {}
for n_clusters in range(2, 9):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(X)
    cluster_results[str(n_clusters)] = labels

# Step 2: Bootstrap the stability
n_rep = 10
n_boot = 100
bootstrap_results = {}
for n_clusters, labels in cluster_results.items():
    stability_scores = []
    for _ in range(n_rep):
        # Bootstrap resampling
        boot_labels = resample(labels, replace=True, random_state=1234)
        stability_scores.append(adjusted_rand_score(labels, boot_labels))
    bootstrap_results[n_clusters] = stability_scores

# Step 3: Plot the global stability boxplot
plt.boxplot(bootstrap_results.values(), labels=list(bootstrap_results.keys()))
plt.xlabel('Number of Segments')
plt.ylabel('Adjusted Rand Index')
plt.title('Global Stability Boxplot')
plt.show()

# Step 4: Plot histogram of a specific segmentation solution (e.g., 4 clusters)
plt.hist(cluster_results['4'], bins=np.arange(5) - 0.5, edgecolor='black')
plt.xlabel('Cluster')
plt.ylabel('Frequency')
plt.title('Histogram of Segmentation Solution (4 clusters)')
plt.show()

# Step 5: Calculate and plot segment level stability
segment_stabilities = []
for n_clusters, labels in cluster_results.items():
    stability_scores = []
    for i in range(int(n_clusters)):
        segment_labels = (np.array(labels) == i).astype(int)
        stability_scores.append(np.mean(segment_labels == resample(segment_labels, replace=True, random_state=1234)))
    segment_stabilities.append(stability_scores)

# Plot segment level stability
for i, scores in enumerate(segment_stabilities):
    plt.plot(scores, label=f'Segment {i + 1}')

plt.xlabel('Segment Number')
plt.ylabel('Segment Stability')
plt.title('Segment Level Stability Within Solutions')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
  # Replace 'load_your_dataset' with the function or method to load your dataset
data = pd.read_excel("C:/Users/Yashwanth G R/Desktop/internship courses/feynn_labs_data.xlsx")
df_num = data.select_dtypes(include = np.number)
df_cat = data.select_dtypes(exclude = np.number)
df_cat_encode = pd.get_dummies(df_cat , drop_first= True)
df = pd.concat([df_num , df_cat_encode], axis = 1)
# Load your dataset
X = df[['convenient_Yes', 'Age']].values # Replace 'load_your_dataset' with the actual method to load your dataset

# Set seed for reproducibility
np.random.seed(1234)

# Step 1: Fit Gaussian Mixture Models with different numbers of components
gmm_models = {}
for n_components in range(2, 9):
    gmm = GaussianMixture(n_components=n_components, random_state=42)
    labels = gmm.fit_predict(X)
    gmm_models[str(n_components)] = (gmm, labels)

# Step 2: Plot the information criteria
aic_values = [gmm_models[key][0].aic(X) for key in gmm_models]
bic_values = [gmm_models[key][0].bic(X) for key in gmm_models]

plt.plot(range(2, 9), aic_values, label='AIC')
plt.plot(range(2, 9), bic_values, label='BIC')
plt.xlabel('Number of Components')
plt.ylabel('Value of Information Criteria')
plt.legend()
plt.show()

# Step 3: Retrieve the model with 4 components
selected_model = gmm_models['4'][0]

# Step 4: Compare clustering results
kmeans_labels = gmm_models['4'][1]

# Print confusion matrix
confusion_matrix = np.zeros((len(np.unique(kmeans_labels)), len(np.unique(selected_model))))
for i, kmeans_label in enumerate(np.unique(kmeans_labels)):
    for j, mixture_label in enumerate(np.unique(selected_model)):
        confusion_matrix[i, j] = np.sum((kmeans_labels == kmeans_label) & (selected_model == mixture_label))

print("Confusion Matrix:")
print(confusion_matrix)
```

```python
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load your dataset using pandas
# Replace 'your_dataset.csv' with the actual file path of your CSV file
df = pd.read_excel('C:/Users/Yashwanth G R/Desktop/internship courses/feynn_labs_data.xlsx')

# Specify the categorical features for modeling
categorical_features = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap', 'tasty', 'expensive', 'healthy', 'disgusting']

# Create numerical representations of categorical features
label_encoder = LabelEncoder()
for feature in categorical_features:
    df[feature + '_encoded'] = label_encoder.fit_transform(df[feature])

# Extract the encoded features
encoded_features = [feature + '_encoded' for feature in categorical_features]

# Fit a Gaussian Mixture Model
gmm = GaussianMixture(n_components=2, random_state=1234)
gmm.fit(df[encoded_features])

# Extract cluster assignments
cluster_assignments = gmm.predict(df[encoded_features])

# Display cluster sizes
print(pd.Series(cluster_assignments).value_counts())

# Plot the results
# Note: You may need to adjust this part based on your specific dataset and features
plt.scatter(df[encoded_features[2]], df[encoded_features[-1]], c=cluster_assignments, cmap='viridis')
plt.title('Gaussian Mixture Model Clustering')
plt.xlabel(encoded_features[2])
plt.ylabel(encoded_features[-1])
plt.show()
```
